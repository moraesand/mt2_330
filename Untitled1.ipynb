# Lecture 9: Classification Metrics

#### Logistic Regression
Logistic Regression is a type of classification algorithm. 
- it computes a weighted sum of input features (like linear regression)
- outputs a probability of it being in one category or another (0 or 1)

#### `predict` vs. `predict_proba`
- `predict` outputs a the predicted category (0 or 1)
- `predict_proba` outputs the probability of either category; 2-column array

#### `C` Hyperparameter
- controls regularization (technique used to reduce overfitting)
    - specifically uses L2 regularization
    - larger `C` = simpler model (stronger regularization)
    - smaller `C` = more complex model (weaker regularization)

#### Coefficients of Features
Each feature will get a weight according to how much the feature pushes the prediction one way or another
- positive (+) coefficient -> pushes prediction towards class 1
- negative (-) coefficient -> pushes prediction towards class 0

***

#### Accuracy, Precision, Recall, & F1-score
Accuracy: the fraction of correct predictions
- can be misleading if classes are imbalanced
    - e.g., if 90% of data is class 0, `DummyClassifier` can get an accuracy of 90% simply by predicting class 0 every time.

Precision: TP / (TP + FP)
- Of all the times the model said "yes," how often was it right?
- High precision means few false alarms.

Recall: TP / (TP + FN)
- Of all the actual "yes" cases, how many did the model catch?
- High recall means you miss fewer positives.

F1-Score: 2 × (precision × recall) / (precision + recall)
- The harmonic mean of both precision & recall; the score given based on those two factors.

*** 

#### EDA (Exploratory Data Analysis) 
A step in the learning process where the data is analyzed, visualized, and cleaned before being put into any model. 
- Visualized via graphs & plots
- Cleaned via clearing missing values, duplicates, class imbalance, etc.
- This step is critical in determining which model to used based on the type of data.

## Lecture 10: Regression Metrics

#### `Ridge` Regression
`RidgeRegression` is a version of `LinearRegressor` that adds L2 Regularization.
- Uses hyperperameter `alpha`:
    - Ridge's version of regularization control
    - higher `alpha` -> higher regularization (more complex model)
    - lower `alpha` -> lower regularization (less complex model)
- In terms of effects, `C` is the inverse of `alpha`
***
#### Regression Metrics
**MSE - Mean Squared Error** 
$$MSE = \frac{1}{n}\sum_{i=1}^n (y_i - \hat{y}_i)^2$$
Things to note: 
- Units are squared, meaning that results can be hard to interpret
- MSE is scale-dependent; cannot compare MSE values across diff. datasets or target scales.

**RMSE - Root Mean Squared Error**
$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$
More interpretable than MSE, but scale-dependency is still an issue.

**R^2 - Coefficient of Determination**
$$R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}$$
Things to note: 
- Max. value is 1.0 - perfect predictions
- R^2 means that it is no better than `DummyRegressor`
- Negative R^2 means that it is worse than `DummyRegressor` - something is really wrong
- Scale-independent; useful for comparing models across different problems
- Default score returned by `model.score()` for regression

**MAPE - Mean Absolute Percent Error**
$$MAPE = \frac{1}{n} \sum_{i=1}^n \left|\frac{y_i - \hat{y}_i}{y_i}\right| \times 100\%$$

MAPE is the most interpretable metric out of the four. Measures avg. % that the prediction is off from the true values.
- MAPE is based of proportionality; relative error matters a lot more than absolute error
    - e.g., if House A was 10,000 w/ RMSE error of 1000 dollars (10% error)
    - House B was 10,000 w/ RMSE error of 5000 dollars (50% error) - House B is 5x worse in relative terms.
- Lower is better; 0% is perfect.

## Lecture 11: Ensembles 
**Ensembles:** combining multiple models to get better predictions than any single model alone. 

#### Random Forests
Random Forest models are decision trees that incorporate randomness in two ways:
- Bootstrap sampling (bagging)
    - Each tree is trained on a different random sample of the training data
- Feature subsampling at each split
    - At every node in the tree, only a random subset of features is considered for the best split (controlled by `max_features`)
    - Prevents every node from using the same feature and offers variety in the overall tree

Two ways prediction works:
- Classification
    - Each tree votes (predicts) a class. Majority wins (first-past-the-post).
- Regression
    - Each tree outputs a number; final prediction is the avg. of those numbers.
 
Hyperperameters:
- `n_estimators`: # of trees.
    - more trees -> more stable predictions, but slower. Vice versa.
    - Unlike most hyperperameters, more trees rarely hurt b/c validation score levels off rather than getting worse.
- `max_depth`: Max. depth of each tree.
    - Deeper trees -> more complex, more prone to overfitting
- `max_features`: # of features to consider at each split.
    - Fewer -> more diverse trees but weaker individual trees.
 
Pros: 
- Strong off-the shelf performance
- No scaling needed
- parallelizable (trees are independent from each other)
- Less prone to overfitting than a single tree

Cons: 
- More memory required
- Slower than one tree
- hard to interpret (more trees = more complex)
- struggles w/ high-dimentional sparse data (like text)

#### Gradient Boosted Trees
Gradient Boosted Trees builds trees sequentially, where each new tree builds off of the previous one and fixes mistakes of the previous. 
- It uses many shallow trees (of depths 1-5) called "weak learners"
- Each tree corrects the residual errors of the current ensemble
- no randomness is applied

Hyperparameters:
- `learning_rate`: controls how aggressively each new tree corrects the previous mistakes.
    - High `learning_rate` -> makes large corrections -> model learns fast but overshoots and overfits.
    - Low `learning_rate` -> make small corrections -> more trees are needed (model is therefore slow) but is more robust.
    - Low `learning_rate` + large # of trees is the usual strat.
- `n_estimators`: # of boosting rounds (more = more complex)
- `max_depth`: depth of each tree
- `scale_pos_weight`: for handling class imbalance
    - balancing of positive and negative weights

There are 3 main libraries for GBT:
- `XGBoost`
- `LightGBM`
- `CatBoost`

All of these are not part of the `sklearn` library but has a similar interface.

Gradient Boosting are typically more accurate than Random Forest but more prone to overfitting if not tuned carefully. 
- Random Forest more robust out of box
- Gradient Boosting is sequential while RFT can be parallelizable.

#### Averaging (`VotingClassifier` / `VotingRegressor`)
Averaging is taking the results of different models to determine a prediction.

`VotingClassifier` takes a list of estimators and combines their predictions at test time.  

There are two types of "voting" (how to estimate):
- **Hard Voting (`voting=hard`):** Each model predicts a class label, and majority wins (first-past-the-post)
- **Soft Voting (`voting=soft`):** Each model outputs a probability (`predict_proba`) and probabilities are averaged across models.
    - This is better for finding confident predictions
 
Pros:
- More models = more accuracy and reduces the chances of error being heightened b/c there are more models
    - only works if error is uncorrelated (if errors are the same, avg. doesen't help)
- Better than using an individual model in terms of accuracy and results

Cons:
- Slow runtime (due to many different models being used)
- Harder to interpret results
- More code to maintain

#### Stacking (`StackingClassifier` / `StackingRegressor`)
Takes the idea of averaging further: instead of averaging the outputs of each estimator, it uses a models outputs as inputs into another model
- Essentially sequential version of averaging

**How it works:**
1. Train base estimators (e.g., logistic regression, random forest, etc.)
    - You can use as many estimators as you want
2. Use cross-validated predictions from base estimators as features for the meta-model
    - Critical: using full training set predictions would cause data leakage b/c base models already seen those examples
    - Logistic regression handles multiple estimators being used
3. Train the meta-model
    - For `StackingClassifier`, logistic regression is used
  
The coefficients of the meta logistic regression tells you how much it trusts each model.
- high positive results: meta-model learned to weight it heavily
- negative results: meta-model learned predictions are anti-correlated with the true prediction (very bad!!)

Pros:
- Learns an optimal weighting instead of applying equal weights to all estimators (like averaging)

Cons:
- Slower
- More complex
- Can overfit if not done carefully.

| Method | Use when... |
|---|---|
| Random Forest | You want a strong, reliable baseline with minimal tuning |
| LightGBM/XGBoost | You want state-of-the-art accuracy and are willing to tune |
| VotingClassifier | You have diverse models and want a quick accuracy boost |
| StackingClassifier | You want maximum accuracy and don't mind complexity/slowness |

## Lecture 12: Feature Importances & Model Transparency

#### SHAP (SHapley Additive exPlanations)
SHAP is a way to explain individual predictions from any model, including complex ones like LightGBM
- It assigns a shapely value (contribution value) to each feature to tell us how much each feature pushed the models prediction one way or another.
- Positive values push the prediction towards outcome 1
- Negative values push the prediction towards outcome 0
- SHAP plots are useful in determining how influential each model was (+ or -) in the outcome of a ensemble-based model.

## Lecture 13: Feature Selection

There are many ways to do feature selection:

**Model-based Selection `SelectFromModel`**  
Use an unsupervised model to rank features by importance, then keep only the ones above a certain threshold.

How it works:
1. Fit model that produces feature importance
2. Discard features that are below a threshold
3. Train your final model on only the selected features

Feature selection model doesn't have to be the final model used. 

Models that can be used: 
- ✅ Random Forest — has `feature_importances_`
- ✅ Logistic Regression / Ridge — has `coef_` (with linear kernel)
- ✅ SVC with linear kernel — has `coef_`
- ❌ KNN — no feature importances
- ❌ SVC with RBF kernel — no `coef_`

**Recursive Feature Elimination (RFE)**  
RFE is more careful than model-based selection. Instead of removing all unimportant features, it removes one at a time, re-fitting after each removal.

How it works:
1. Decide k (how many features you want to end up with)
2. Fit model on all current features
3. Find least important feature
4. Remove it
5. Repeat until k features remain

This is good b/c the model evaluates after each removal to see whether unassuming features that were removed were actually important (due to correlation)
- Feature importance is context-dependent

`RFECV` is a version of RFE that automatically selects k based on running CV at each step and selecting the k that gave the best CV score. 
- Trade-off is speed: CV inside CV, which can be time consuming.

**Forward Selection**  
Instead of starting w/ features and removing, forward selection starts w/ no features and adds them one at a time.

How it works:
1. Starts w/ empty feature set
2. Evaluates each remaining feature one at a time, evaluate CV score each time
3. Adds whichever feature improves score the most
4. Repeat until adding more features doesn't help

Also known as a wrapper method, forward selection is a greedy approximation that doesn't guarantee finding the most optimal, but very close in practice. 
- Some cons is that it might end up with suboptimal feature sets.
